```python
data.groupby(by="Pclass")["Age"].mean()
```

```python
def remp_age(data):
    # Calculer la moyenne de l'âge par classe
    mean_age_by_class = data.groupby(by="Pclass")["Age"].mean()
    
    # Remplacer les valeurs manquantes par la moyenne de chaque classe
    for i in range(data.shape[0]):
        if pd.isna(data.loc[i, "Age"]):
            data.loc[i, "Age"] = int(mean_age_by_class[data.loc[i, "Pclass"]])
```

```python
remp_age(data)
```
